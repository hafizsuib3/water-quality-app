<font color="ORANGE" size="5px"> DATA CLEANING - ANALYZE DATA BEHAVIOR USING A HISTOGRAM, IDENTIFY OUTLIERS, AND REMOVE THEM </font>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to replace outliers using the median and plot histograms
def plot_and_replace_outliers_with_median(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)

    # Filter for numeric columns only
    numeric_cols = df.select_dtypes(include=['number']).columns

    # Create a copy of the DataFrame for cleaning
    cleaned_df = df.copy()

    # Iterate over each numeric column for analysis and replacement
    for col in numeric_cols:
        # Calculate Q1, Q3, and IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Calculate the median of the column for non-outlier data
        median_value = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)][col].median()

        # Replace outliers with the median
        outliers = (df[col] < lower_bound) | (df[col] > upper_bound)
        cleaned_df.loc[outliers, col] = median_value

        # Plotting
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)  # Histogram before replacement
        df[col].hist(bins=30, edgecolor='black')
        plt.title(f'{col} - Before Replacement')

        plt.subplot(1, 2, 2)  # Histogram after replacement
        cleaned_df[col].hist(bins=30, edgecolor='black')
        plt.title(f'{col} - After Replacement')
        plt.tight_layout()
        plt.show()

        print(f"Replaced outliers in {col} with median value {median_value}.")

    # Return the cleaned DataFrame
    return cleaned_df

# Specify your file path
file_path = '/content/drive/MyDrive/UTM IN GD/SEM 3/THESIS/DATA/india/SW_CWC_UttarPradesh2007-2021-SELECTEDPARAMETERS-REMOVEDCOLORODOR-DATACLEANED.csv'
clean_df = plot_and_replace_outliers_with_median(file_path)

# Save the cleaned DataFrame to a new CSV file
cleaned_file_path = '/content/drive/MyDrive/UTM IN GD/SEM 3/THESIS/DATA/india/SW_CWC_UttarPradesh2007-2021-SELECTEDPARAMETERS-REMOVEDCOLORODOR-DATACLEANED-OUTLIERSREMOVED.csv'
clean_df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")


```

<font color="GREEN" size="5px"> WQI CALCULATION - CALCULATING WQI USING WEIGHTED PARAMETERS </font>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier

# Load the new data file
file_path = '/content/drive/MyDrive/UTM IN GD/SEM 3/THESIS/DATA/india/SW_CWC_UttarPradesh2007-2021-SELECTEDPARAMETERS-REMOVEDCOLORODOR-DATACLEANED-OUTLIERSREMOVED.csv'
df = pd.read_csv(file_path)

# Define permissible limits and ideal values for WQI calculation
permissible_limits = {
    'BIOCHEMICAL OXYGEN DEMAND(mg/L)': 3,
    'PH': 8.5,
    'DISSOLVED OXYGEN(mg/L)': 5,
    'NITROGEN, TOTAL OXIDISED(NO2+NO3)': 45,
    'ELECTRICAL CONDUCTIVITY(µS/CM) at 25°': 300,
    'TOTAL COLIFORMS(MPN/100ml)': 2500,
    'FECAL COLIFORMS(mpn/100ml)': 500
}
ideal_values = {
    'BIOCHEMICAL OXYGEN DEMAND(mg/L)': 0,
    'PH': 7.0,
    'DISSOLVED OXYGEN(mg/L)': 14.6,
    'NITROGEN, TOTAL OXIDISED(NO2+NO3)': 0,
    'ELECTRICAL CONDUCTIVITY(µS/CM) at 25°': 0,
    'TOTAL COLIFORMS(MPN/100ml)': 0,
    'FECAL COLIFORMS(mpn/100ml)': 0
}

# Calculate WQI
K = 1 / sum(1 / Si for Si in permissible_limits.values())
weights = {param: K / Si for param, Si in permissible_limits.items()}
def calculate_wqi(row):
    wqi = 0
    for param, Si in permissible_limits.items():
        if param in row:
            vi = row[param]
            v0 = ideal_values[param]
            Qi = 100 * (vi - v0) / (Si - v0) if Si != v0 else 0
            Wi = weights[param]
            wqi += (Qi * Wi)
    return wqi / sum(weights.values())
df['WQI'] = df.apply(calculate_wqi, axis=1)

# Save the WQI results to a new CSV file
new_file_path = '/content/drive/MyDrive/UTM IN GD/SEM 3/THESIS/DATA/india/SW_CWC_UttarPradesh2007-2021-SELECTEDPARAMETERS-REMOVEDCOLORODOR-DATACLEANED-OUTLIERSREMOVED-WQI.csv'
df.to_csv(new_file_path, index=False)
print("WQI calculated and saved to new CSV file.")



```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier

# Define the class mapping
class_mapping = {0: 'B(Good)', 1: 'C(Poor)', 2: 'D(Very Poor)', 3: 'E(Unsuitable for drinking)'}

# Define the models
models = {
    "SVM": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "MLP": MLPClassifier(hidden_layer_sizes=(100, 100, 100))
}

# Let's assume X_train_smote, y_train_smote, X_test, y_test are already defined and appropriate
# Fit each model, predict and generate confusion matrix, and plot
for name, model in models.items():
    model.fit(X_train_smote, y_train_smote)
    predictions = model.predict(X_test)

    # Generate confusion matrix with class mapping
    cm = confusion_matrix(y_test, predictions, labels=list(class_mapping.keys()))
    df_cm = pd.DataFrame(cm, index=[class_mapping[i] for i in model.classes_], columns=[class_mapping[i] for i in model.classes_])

    # Plotting the confusion matrix
    plt.figure(figsize=(10, 7))
    ax = sns.heatmap(df_cm, annot=True, fmt="d", cmap='Blues', cbar_kws={'label': 'Frequency'})
    plt.title(f'{name} - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')

    # Calculating accuracy and misclassification
    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy
    plt.figtext(0.5, -0.1, f'Accuracy={accuracy:.4f}; Misclass={misclass:.4f}', ha='center', va='center', fontsize=12)
    plt.show()






```
